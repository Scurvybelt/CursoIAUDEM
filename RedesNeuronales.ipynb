```python
# Red Neuronal Artificial (ANN)
import pandas as pd
import numpy as np

df = pd.read_csv("diabetes2.csv")
df
```

```python
# Configuracion de RNA (Red Neuronal Artificial) para realizar prediccion
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

# Dividimos datos y seleccionamos variables
X = df.drop('Outcome', axis=1).values # Obtener un Array de Numpy
y = df['Outcome'].values # Obtener un Array de Numpy

# la red neuronal solo debe recibir vectores de datos, no dataframes!
# el algoritmo si puede recibir dataframes, no necesariamente vectores

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)

```
